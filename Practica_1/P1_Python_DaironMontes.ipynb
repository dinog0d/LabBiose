a. Cree el siguiente par de vectores:
a = [67.1,1, -0.3, 5.2, -6], b= [1, 3, 2.2, 5.1, 1]

```python
a = np.array([67.1,1,-0.3,5.2,-6])
b = np.array([1,3,2.2,5.1,1])

print(a,b)
```

b. Implemente la multiplicaci√≥n escalar de a.b ¬øQu√© se debe realizar para poder multiplicar los dos
vectores?

- Para realizar esta multiplicaci√≥n, debemos primero asegurarnos que los vectores a utilizar tengan la misma dimensi√≥n, luego de esto, el procedimiento es bastante sencillo: Se multiplica cada componente del vector a por el componente correspondiente del vector b, una vez realizado esto, se suman todos los productos obtenidos en el paso anterior. Esta suma dar√° como resultado un valor escalar.

```python
# Este prodecimiento se puede realizar con funciones de la propia librer√≠a de Numpy

print("Forma y dimensi√≥n de a:",np.shape(a),np.ndim(a))

print("Forma y dimensi√≥n de b:",np.shape(b),np.ndim(b))

producto_escalar = np.dot(a,b)
print(producto_escalar)
```

c. Implemente la multiplicaci√≥n punto a punto de a.b.

```python
# Este prodecimiento se puede realizar con funciones de la propia librer√≠a de Numpy

producto_punto = np.multiply(a,b)
print(producto_punto)
```

f. Consulte la funci√≥n de los siguientes comandos de la librer√≠a numpy:
- ones
- round
- ceil
- floor

e implemente cada funci√≥n para ejemplificar su uso.

```python
# Seg√∫n la documentaci√≥n de Numpy oficial, existen diferentes formas de crear y modificar matematicamente arrays de Numpy
# Algunos casos especiales y necesarios para el uso de la liber√≠a son los siguientes

# Si se desea crear un array completo de '1's

Matriz_de_unos = np.ones(11) # El valor dentro de la funci√≥n np.ones(_) indica la forma o longitud deseada del array (o cuantos '1's deseamos)
print("np.ones:",Matriz_de_unos)

# Al tener un array de flotantes, podemos redondearlos parejamente y seg√∫n la cantidad de decimales que deseemos

pi = np.ones(2)*np.pi # Usaremos la funci√≥n anterior para crear un array con el valor de pi

# Creado este array, usamos la funci√≥n np.round para observar este valor con solo 3 decimales (decimales de ejemplo)

round_pi = np.round(pi,3)
print("np.round:",round_pi)

# As√≠ mismo tenemos alternativas para la funci√≥n anterior, dependiendo si queremos redondear hacia un valor superior o un valor inferior al que tenemos
# As√≠ como dice la documentaci√≥n: El valor 'ceil' (techo) del escalar x es el entero m√°s peque√±o i, tal que i >= x.

# Ejemplo de la documentaci√≥n:

a = np.array([-1.7, -1.5, -0.2, 0.2, 1.5, 1.7, 2.0])
a_techada = np.ceil(a)
print("np.ceil:",a_techada)

# En contraste a la funci√≥n np.ceil, el valor 'floor' (piso) del escalar x es el entero m√°s grande i, tal que i <= x.

# Ejemplo de la documentaci√≥n:

b = np.array([-1.7, -1.5, -0.2, 0.2, 1.5, 1.7, 2.0])
b_pisada = np.floor(b)
print("np.floor:",b_pisada)
```

j. Construya la funci√≥n ùë¶[ùëõ] = sin (ùúã ‚àó 0.18ùëõ) en el intervalo 0 <= n <= 80.


```python
# Para construir esta funci√≥n y, podemos usar la funci√≥n np.sin de Numpy, esta nos permite crear una funci√≥n seno que recibe de parametros un valor o un array
x = np.arange(0,81) # Nos apoyamos de la funci√≥n arange para crear un array con los valores de 0 a 80
y = np.sin ( x * 0.18 * np.pi)

print("Valores de la funci√≥n y[n]:",y)
```

k. Construya otra senoidal ùë¶2[ùëõ] = cos(2ùúã ‚àó 0.03ùëõ)

```python
# Nos apoyamos del anterior array con los valores de 0 a 80 para crear la funcion

y2 = np.cos ( x * 2 * np.pi * 0.03)

print("Valores de la funci√≥n y2[n]:",y2)
```

```python
# Ya que ambas se apoyan en el mismo array de rango 0 a 80, significa que ambas funciones tienen la misma dimensi√≥n y forma, por lo tanto, podemos realizar operaciones sencillas entre ellas
# Esto es posible gracias a que los arrays operan elemento a elemento

s = y + y2
print("Valores de la funci√≥n s[n]:",s)

t = y * y2
print("Valores de la funci√≥n t[n]:",t)
```
