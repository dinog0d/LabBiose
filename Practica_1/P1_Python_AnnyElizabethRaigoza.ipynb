**RETO**

Escribir una funciÃ³n que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota mÃ­nima, la mÃ¡xima, media y  la desviaciÃ³n tÃ­pica.

**SoluciÃ³n**
Para especificar el reto se elaborÃ³ un diccionario que contiene las notas finales de 5 alumnos, una vez construido, se enviaron los datos a la funciÃ³n evaluar_notas en la cual se convirtiÃ³ el diccionario a una serie con el objetivo de realizar sus respectivos cÃ¡lculos estadÃ­sticos, donde por medio de la funciÃ³n **.min** se calcula el mÃ­nimo de un arreglo de nÃºmeros, con **.max** se evalua el mÃ¡ximo entre un conjunto de elementos, con **.mean** se encuentra el valor medio de los datos y finalmente con la funciÃ³n **.std** se plantea la desviaciÃ³n de los valores.
Una vez calculadas las estadÃ­stivas se retornan como otro arreglo unidimensional (serie), para luego ser mostrados en consola.

```python
def evaluar_notas(diccionario_estudiantes):
    serie = pd.Series(diccionario_estudiantes)
    minimo = serie.min()
    maximo = serie.max()
    media = serie.mean()
    desviacion = serie.std()
    return pd.Series([minimo, maximo, media, desviacion], index=['minimo', 'maximo', 'media', 'desviacion'])

```

```python
diccionario_estudiantes = {
        'Estudiante 1': 3.5,
        'Estudiante 2': 4.0,
        'Estudiante 3': 2.5,
        'Estudiante 4': 4.5,
        'Estudiante 5': 3.0
}

notas_estudiantes = evaluar_notas(diccionario_estudiantes)
print(notas_estudiantes)
```
