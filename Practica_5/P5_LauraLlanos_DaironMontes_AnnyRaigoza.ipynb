````python
import matplotlib.pyplot as plt
import numpy as np

Fo = 40  # Frecuencia fundamental de la señal
Tp = 1/Fo  # Periodo de la señal
Fs = 1000  # Frecuencia de muestreo
T = 1/Fs  # Periodo de muestreo
t = np.arange(0, Tp+T, T)  # Tiempo para un ciclo de la señal
A = 5
x = A*np.sin(2*np.pi*Fo*t)

fo = Fo/Fs
n = np.arange(0, len(t))
x1 = A*np.sin(2*np.pi*fo*n)

# Calcular la diferencias entre las dos señales
diferencia = np.abs(x[:5] - x1[:5])
print("Diferencias entre las primers 5 muestras:", diferencia)

# Graficas de la señal
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, x, marker='o', label='Señal continua',color='#88B04B')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Señal Continua')
plt.legend()

plt.subplot(2, 1, 2)
plt.stem(n, x1, label='Señal discreta',linefmt='#FF6F61')
plt.xlabel('Muestras (n)')
plt.ylabel('Amplitud')
plt.title('Señal Discreta')
plt.legend()

plt.tight_layout()
plt.show()
````

````python
#Determinar si la potencia es la misma en x(n) tomando 1 o 10 ciclos para la señal continua y discreta

# Determinar la potencia para 1 ciclo, Señal Continua
energia1 = sum(x**2)
potencia1 = energia1/(len(t)-1)
rms1 = np.sqrt(potencia1)

# Determinar la potencia para 10 ciclos, Señal Continua
tiempo10c = np.arange(0,10*Tp+T,T)
x10 = A*np.sin(2*np.pi*Fo*tiempo10c)
energia10 = sum(x10**2)
potencia10 = energia10/(len(tiempo10c)-1)
rms10 = np.sqrt(potencia10)

# Determinar la potencia para 1 ciclo, Señal Discreta
energia1_d = sum(x1**2)
potencia1_d = energia1_d/(len(n)-1)
rms1_d = np.sqrt(potencia1_d)


#  Determinar la potencia para 10 ciclos, Señal Discreta
n10 = np.arange(0, 10*len(t)) # Tomando 10 muestras en un periodo
x10_d = A * np.sin(2 * np.pi * fo * n10)
energia10_d = sum(x10_d**2)
potencia10_d = energia10_d / (len(n10)-1)
rms10_d = np.sqrt(potencia10_d)


print(f"Potencia para 1 ciclo (continua): {potencia1}")
print(f"Potencia para 10 ciclos (continua): {potencia10}")
print(f"Potencia para 1 ciclo (discreta): {potencia1_d}")
print(f"Potencia para 10 ciclos (discreta): {potencia10_d}")

# Comparar los valores de potencia
tolerancia = 0.01

if abs(potencia1 - potencia10)< tolerancia :
  print("La potencia es similar para 1 y 10 ciclos de la señal continua.")
else:
  print("La potencia NO es similar para 1 y 10 ciclos de la señal continua.")


if abs(potencia1_d - potencia10_d)<tolerancia:
  print("La potencia es similar para 1 y 10 ciclos de la señal discreta.")
else:
  print("La potencia NO es similar para 1 y 10 ciclos de la señal discreta.")
````

*b) La potencia es la misma que en 𝑥(𝑛). ¿Por qué?*
Por teoría, para una señal senoidal, la potencia media es constante, independientemente de los ciclos a considerar debido a la naturaleza de la señal, cuya característica principal es la periodicidad, por ello,  características como la energía y potencia se repiten cada cierto periodo; lo cual, se traduce a que la relación de energía a muestras se mantiene constante y, en consecuencia, su potencia también.

En otras palabras, la potencia media se mantiene igual porque se está promediando sobre un mayor número de muestras. La cantidad total de energía en 10 ciclos es básicamente 10 veces la energía de un ciclo, pero dividiendo su total por 10 veces más muestras. **I.** Cree una señal que sea la suma de tres componentes sinusoidales con frecuencias de 40,
80 y 160 Hz.

Defina la frecuencia de muestreo mínima necesaria para representar la señal,
y utilice la frecuencia de muestreo necesaria para representarla apropiadamente

**₊˚ ‿︵‿︵‿︵୨୧ · · ♡ · · ୨୧‿︵‿︵‿︵ ˚₊**                                                                
El Teorema de Nyquist establece que, para reconstruir una señal sin distorsión a partir de muestras digitales, la frecuencia de muestreo (número de muestras por segundo) debe ser al menos el doble de la frecuencia máxima (más alta) contenida en la señal original / de la información que hay, es decir, el periodo de muestreo debe ser por lo menos la mitad del periodo de las oscilaciones más rápidas.
Además, se recomienda observar la periodicidad de la onda, dado que los modelos se construyen con base en la memoria del sistema. [1]

Para evitar aliasing, dado que la frecuencia más alta en la señal es 160 Hz, la frecuencia de muestreo mínima sería: 𝑓 muestreo ≥ 2 × 160 = 320 Hz

````python
import matplotlib.pyplot as plt
import numpy as np
from numpy import sin as sen

fs = 160 * 200  # Frecuencia de muestreo (Hz)
fo = 160  # Frecuencia fundamental (Hz)
T = 1 / fo  # Período fundamental (segundos)
T_muestreo = 1 / fs  # Período de muestreo (segundos)

t = np.arange(0, 40 * T+T_muestreo, T_muestreo)  # Vector de tiempo donde se observan 10 ciclos

f1 = 40  # Frecuencia del primer componente sinusoidal (Hz)
f2 = 80  # Frecuencia del segundo componente sinusoidal (Hz)
f3 = 160  # Frecuencia del tercer componente sinusoidal (Hz)

x1 = sen(2 * np.pi * f1 * t)
x2 = sen(2 * np.pi * f2 * t)
x3 = sen(2 * np.pi * f3 * t)

x = x1 + x2 + x3  # Señal resultante (suma de las tres componentes)

plt.figure(figsize=(10, 6))
plt.plot(t, x1, label='Componente 1', color='#FF69B4')
plt.plot(t, x2, label='Componente 2', color='#40E0D0')
plt.plot(t, x3, label='Componente 3',color='#FF7F50')
plt.legend()
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Señal con tres componentes sinusoidales')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(t, x,color='deeppink')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Señal con tres componentes sinusoidales sumados')
plt.grid(True)
plt.show()
````

**II.** Calcule la transformada de Fourier de la señal y grafique el espectro de frecuencia.


**₊˚ ‿︵‿︵‿︵୨୧ · · ♡ · · ୨୧‿︵‿︵‿︵ ˚₊**                                                                
*Identifique en el espectro las frecuencias que componen la señal (trabaje con 10 periodos de la señal).*

Sabemos que los picos en la gráfica de magnitud indican las frecuencias presentes en la señal [2], teniendo en cuenta que nuestra señal se compone de tres sinoidales con frecuencias de 40, 80 y 160 Hz, podemos evidenciar gráficamente que los picos se encuentran en:(40,4001.17),(79.99,4000.21) y (159.98,3999.15)

Lo cual nos indica que el espectro se compone de:

✬ Una frecuencia fundamental a 40 Hz.

✬ Su armónica a 80 Hz (2 veces la frecuencia fundamental).

✬ Otra armónica a 160 Hz (4 veces la frecuencia fundamental).


````python
# Calcula la transformada de Fourier de la señal
X = np.fft.fft(x)
magnitud = np.abs(X)

N = len(X)
F = np.arange(0,N)*fs/N

plt.figure(figsize=(10, 6))
plt.plot(F,abs(X), color='royalblue')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.title('Espectro de frecuencia de la señal')
plt.grid(True)
plt.show()

# Gráfica con limitación en el eje x
plt.figure(figsize=(10, 6))
plt.plot(F, magnitud, color='royalblue')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.title('Espectro de frecuencia de la señal')
plt.xlim(0, 200)
plt.grid(True)
plt.show()
````

````python
# Identifica las frecuencias en el espectro
# Busca los picos más altos en el espectro y encuentra sus frecuencias correspondientes.
picos = []
magnitud = abs(X)
for i in range(len(magnitud)):
  if magnitud[i] > 0.5 * np.max(magnitud):  # Ajustamos el umbral para identificar los picos relevantes.
    picos.append((F[i], magnitud[i]))


plt.figure(figsize=(10, 6))
plt.plot(F,abs(X))
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.title('Espectro de frecuencia de la señal')
plt.xlim(0,200)
plt.grid(True)

print("Frecuencias identificadas:")
for f, mag in picos[0:3]:
  print(f"Frecuencia: {f:.2f} Hz, Magnitud: {mag:.2f}")
  plt.axvline(x=f, color='red')
  plt.text(f, np.max(abs(X)) * 0.9, str(round(f)), color='red')

plt.show()
````

Las líneas mencionadas anteriormente aplican un filtro pasa-banda. Al realizar la FFT de la señal en la primera línea que convierte la señal del dominio del tiempo al de la frecuencia, podemos preservar un pequeño rango de coeficientes de frecuencia de F1 en f2 en la línea 3, haciendo que las demás componentes de frecuencia sean eliminadas.
Recordemos que un filtro puede definirse como un circuito o sistema selectivo en frecuencia que se emplea para eliminar cierto intervalo de frecuencias [3], en este proceso, vemos que la eliminación de algunas frecuencias y el paso de un conjunto límitado de otras describe el comportamiento de un filtro.

Posteriormente, se ejecuta una línea que realiza la transformada inversa de Fourier (IFFT), para convertir la señal modificada F2 de vuelta al dominio del tiempo, obteniendo una señal senoidal filtrada compuesta principalmente de las frecuencias seleccionadas en los índices 9 a 12.

Extraiga de manera similar cada una de las componentes de la señal.

````python
import numpy as np
import matplotlib.pyplot as plt

# FFT de la señal
F1 = np.fft.fft(x)

# Vector de frecuencias correspondiente
frecuencias = np.fft.fftfreq(len(x), d=1/fs)

# Índices correspondientes a 40 Hz, 80 Hz, y 160 Hz
indice_40 = np.argmin(np.abs(frecuencias - 40))
indice_80 = np.argmin(np.abs(frecuencias - 80))
indice_160 = np.argmin(np.abs(frecuencias - 160))

# Filtramos cada componente por separado
def filtrar_componente(indice):
    F2 = np.zeros(len(F1), dtype=complex)
    F2[indice] = F1[indice]  # Asignamos solo la frecuencia deseada
    F2[-indice] = F1[-indice]  # Simétrica para frecuencias negativas
    return np.fft.ifft(F2)

# Señales filtradas para 40 Hz, 80 Hz y 160 Hz
xr_40 = filtrar_componente(indice_40)
xr_80 = filtrar_componente(indice_80)
xr_160 = filtrar_componente(indice_160)

# Graficamos cada una de las señales filtradas
fig, axs = plt.subplots(3, 1, figsize=(8, 6))

axs[0].plot(t, np.real(xr_40),color='#92A8D1')
axs[0].set(title='Componente 40 Hz', xlabel='Tiempo (s)', ylabel='Amplitud (V)')

axs[1].plot(t, np.real(xr_80),color='#B565A7')
axs[1].set(title='Componente 80 Hz', xlabel='Tiempo (s)', ylabel='Amplitud (V)')

axs[2].plot(t, np.real(xr_160),color='#009B77')
axs[2].set(title='Componente 160 Hz', xlabel='Tiempo (s)', ylabel='Amplitud (V)')

plt.tight_layout()
plt.show()
````


**₊˚ ‿︵‿︵‿︵୨୧ · · ♡ · · ୨୧‿︵‿︵‿︵ ˚₊**    
La densidad espectral de potencia proporciona una representación de cómo la potencia de una señal está distribuida en las diferentes frecuencias, para lograr esto, se puede usar el método de Welch, que consiste en dividir la señal en segmentos superpuestos que en nuestro caso, cada segmento se le aplica una ventana de Hanning para minimizar el efecto de discontinuidades en los extremos de las secciones tomadas.

Por otra parte se resalta que el solapamiento es la cantidad de muestras que las ventanas consecutivas de la señal comparten. En consecuencia, un mayor solapamiento ayuda a tener una mejor estimación de la densidad espectral de potencia porque proporciona más datos para cada ventana, reduciendo la varianza entre estas. Es recomendado utilizar un solapamiento del 50% en adelante [4, 5] para tomar más fragmentos de la señal en diferentes momentos y así, poder analizar qué está pasando en las distintas frecuencias de la señal. Por otro lado, si las partes de la onda analizada no se superponen, el análisis es tan rápido que se pueden perder detalles importantes.

Adicionando a lo anterior, el tamaño de la ventana nos indica la cantidad de muestras utilizadas en cada segmento para calcular el espectro. Por ello, se recomienda que su tamaño sea grande [5,6], para poder capturar más parte de la señal en cada paso, esto, nos permite ver mejor los detalles en las frecuencias, es decir, es posible distinguir frecuencias cercanas entre sí, siempre y cuando la señal no varíe demasiado con el tiempo, porque de lo contrario, no se puede detectar bien los cambios rápidos.
Caso contrario si se usa una ventana pequeña, la cual, nos permite ver mejor los cambios rápidos en la señal y tener una mejor resolución en el tiempo. Sin embargo, la desventaja es que se pierde precisión para ver frecuencias cercanas.

Es así como, para nuestro caso, se usó un solapamiento del 50% (valor noverlap=128) para una ventana de tamaño 256 muestras. Dado que, en señales como el ECG, que pueden tener variaciones importantes tanto en frecuencia como en tiempo.

````python
import os
import scipy.io as sio
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np

# Definir rutas
control_path = '/content/Lab_Biose_Dinoequipo/Practica_3/control/'
parkinson_path = '/content/Lab_Biose_Dinoequipo/Practica_3/parkinson/'

# Listar los archivos en las carpetas
directoriosControl = os.listdir(control_path)
directoriosPark = os.listdir(parkinson_path)

# Cargar las señales EEG (ajusta según el archivo específico que necesites)
control_file = control_path + directoriosControl[0]  # Archivo de control
parkinson_file = parkinson_path + directoriosPark[1]  # Archivo de Parkinson

# Cargar los datos desde los archivos .mat
control_data = sio.loadmat(control_file)
parkinson_data = sio.loadmat(parkinson_file)

# Obtener las señales del canal deseado (ejemplo: canal 1 para ambos grupos)
control_signal = control_data['data']
parkinson_signal = parkinson_data['data']

# Dimensiones de las señales
canalesC, muestrasC, epocasC = control_signal.shape
canalesP, muestrasP, epocasP = parkinson_signal.shape

# Aplanar las señales para que sean continuas
control_signal = np.reshape(control_signal, (canalesC, muestrasC * epocasC), order="F")
parkinson_signal = np.reshape(parkinson_signal, (canalesP, muestrasP * epocasP), order="F")

# Graficar las señales EEG del paciente sano y con Parkinson
fig, axs = plt.subplots(2, 1, figsize=(10, 6))
axs[0].plot(control_signal[1, :], color='#FF6F61')
axs[0].set_title('Señal EEG del paciente Sano')
axs[0].set_xlabel('Muestras')
axs[0].set_ylabel('Amplitud')
axs[0].set_xlim([0, 2000])
axs[0].grid(True)

axs[1].plot(parkinson_signal[1, :], color='#2ECC71')
axs[1].set_title('Señal EEG del paciente con Parkinson')
axs[1].set_xlabel('Muestras')
axs[1].set_ylabel('Amplitud')
axs[1].set_xlim([0, 2000])
axs[1].grid(True)

plt.tight_layout()
plt.show()

# Parámetros para el Periodograma de Welch
fs = 1200  # Frecuencia de muestreo ajustada según tu contexto
nperseg = 1200  # Ancho de la ventana (potencia de 2 para eficiencia)
noverlap = int(nperseg / 2) # 50% de solapamiento para un buen equilibrio entre resolución y varianza.

# Calcular el periodograma de Welch para ambas señales
f_control, Pxx_control = signal.welch(control_signal[1, :], fs, window='hann', nperseg=nperseg, noverlap=noverlap)
f_parkinson, Pxx_parkinson = signal.welch(parkinson_signal[1, :], fs, window='hann', nperseg=nperseg, noverlap=noverlap)

# Graficar los periodogramas de Welch
fig, axs = plt.subplots(2, 1, figsize=(10, 6))

axs[0].stem(f_control, Pxx_control, linefmt='#3498DB', basefmt=" ")
axs[0].set_ylabel('Potencia (V^2/Hz)')
axs[0].set_xlabel('Frecuencia (Hz)')
axs[0].set_title("Periodograma de Welch - sujeto control")
axs[0].set_xlim(0, 70)
axs[0].grid(True)

axs[1].stem(f_parkinson, Pxx_parkinson, linefmt='#9B59B6', basefmt=" ")
axs[1].set_ylabel('Potencia (V^2/Hz)')
axs[1].set_xlabel('Frecuencia (Hz)')
axs[1].set_title("Periodograma de Welch - sujeto con Parkinson")
axs[1].set_xlim(0, 70)
axs[1].grid(True)

plt.tight_layout()
plt.show()

fig = plt.figure(figsize=(10, 6))
plt.stem(f_control, Pxx_control, label='Control', linefmt='#FF3D00')
plt.stem(f_parkinson, Pxx_parkinson, label='Parkinson', linefmt='#FF66B2')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Potencia (V^2/Hz)')
plt.title('Periodograma de Welch')
plt.xlim(0, 70)
plt.legend()
plt.grid(True)
plt.show()
````

**JUSTIFICACIÓN DE ELECCIÓN DE PARÁMETROS**

Para el proceso de análisis frecuencial de señales fisiológicas la elección adecuada del solapamiento y el ancho de ventana en el método de Welch es una etapa fundamental para la obtención de una estimación precisa de la densidad espectral de potencia, debido a que estos factores tienen una influencia directa en la calidad de la resolución frecuencial y las variaciones de la estimación.

**nperseg= Ancho de ventana:**

Con el fin de preservar la eficiencia computacional se seleccionó un ancho de ventana de 1200 muestras hecho que equivale a una potencia de trabajo igual a dos, esto también se elige teniendo en cuenta que la transformada de Fourier presenta sus mayores rangos de eficiencia en las potencias de dos.
Por otra parte, al trabajar con un ancho de ventana más amplio se puede adquirir una mayor resolución frecuencial sacrificando estabilidad, ya que la varianza de la estimación incrementa paulatinamente respecto al crecimiento de la ventana. Teniendo en cuenta este hecho se optó por la implementación de un valor de 1200 como un dato que equipara de forma razonable este fenómeno y que considera la longitud de la señal EEG y la frecuencia de muestreo.

**noverlap= Solapamiento:**

Teniendo como objetivo la reducción de la varianza en la estimación y el mejoramiento de la resolución en frecuencias bajas, se implementó un solapamiento del 50%, el cual indica que cada mitad del segmento se superpone entre sí con otra sección tomada durante el procesamiento. Este porcentaje se establece como una media estándar que proporciona una adecuada relación entre el gasto computacional y la reducción de la varianza.

**ANÁLISIS DEL PERIODOGRAMA DE WELCH DEL PACIENTE DE CONTROL Y EL SANO**

Por medio del periodograma de Welch, se evidenció la presencia de picos iniciales pronunciados en un rango de frecuencias de 0 a 2 Hz para ambos individuos analizados (paciente de control y paciente sano), los cuales decrecen en su amplitud rápidamente a partir de los 8 y 10 Hz, indicando una actividad básica común. En este punto es importante resaltar que la caída de potencia es menos pronunciada en el sujeto diagnosticado con la enfermedad de Parkinson siendo su intervalo de potencias mayor respecto al sujeto de control.

Adicionalmente, al observar el comportamiento del periodograma del sujeto de control se indica su veloz disminución de potencia entre un rango de 10 a 30Hz llegando a ser extremadamente baja (casi imperceptible a simple vista) al sobrepasar los 15 Hz, sin embargo, al visualizar el comportamiento de la señal del individuo con Parkinson se indica una tendencia sostenida y con mayor amplitud en contraposición con la señal de control, especialmente en frecuencias de 10 a 20Hz, siendo incluso posible su visualización en valores frecuenciales más elevados, este tipo de comportamientos indica una mayor actividad y presencia de la enfermedad en intervalos de frecuencia más elevados, no obstante, al sobrepasar los 30 Hz se identifica que las potencias son casi nulas y la información contenida en esas secciones es irrelevante.

Los incrementos de potencia del paciente con Parkinson en los rangos bajos de frecuencia pueden encontrarse relacionados con las diversas alteraciones motoras asociadas a la enfermedad, cuya suavidad en la caída de la amplitud señala una mayor oscilación en este rango de frecuencias específicas, este planteamiento se constata en el descenso rápido de la potencia del individuo sano al superar los 3Hz, fenómeno que puede indicar la manifestación de una señal más estable, constante y menos ruidosa en individuos sanos.

Todo lo anterior puede ser reafirmado gracias al artículo “EEG spectral and connectivity parameters as cognitive biomarkers in Parkinson disease”,  a través del cual se indica que los pacientes diagnosticados con la enfermedad de Parkinson presentaron una potencia significativamente mayor en el análisis de todas las bandas frecuenciales de la región temporal derecha frente a sujetos de control sanos, en el artículo concluyen que las medidas de la potencia relativa espectral son un complemento efectivo para la evaluación en el área de la neuropsicología, permitiendo examinar el deterioro cognitivo relacionado con la enfermedad [7].
