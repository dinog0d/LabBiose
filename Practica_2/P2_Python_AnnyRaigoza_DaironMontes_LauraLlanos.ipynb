Implemente una funci√≥n en Python que permita generar una funci√≥n escal√≥n unitario, definida en
un intervalo ùëõ1‚â§ùëõ0‚â§ùëõ2

```python
#Importamos las librer√≠as que necesitemos para operar
import numpy as np
import matplotlib.pyplot as plt

#Hacemos una funci√≥n que permita el ingresar √∫nicamente n√∫meros para evitar errores
def verificacionEntero(a):
    while not a.isnumeric():
      print("Ha ingresado un car√°cter no v√°lido")
      a = input("Ingrese un valor num√©rico: ")
    return (a)

#Hacemos una funci√≥n que nos permita crear el vector tiempo
def vectorEscalon(t,duracion,muestreo):
  escalon = list(np.arange(0, t, muestreo))
  muestreo = 0.01

  i = 0  #Dado que en el escal√≥n es 0 si n<n0, por eso iniciamos 'i' en cero.
  while i<len(escalon):
    if i<t:
      escalon.append(0)
      i += 1
    if i>=t:
      escalon.append(1)
      i += 1
  return escalon

#Solicitamos el intervalo en el que desea el escal√≥n unitario
n0=input("Ingrese el tiempo (en segundos) en el que desea que inicie su escal√≥n unitario: ")
verificacionEntero(n0)
x=int(n0)
duracion=input("Ingrese la duraci√≥n total del escal√≥n (en segundos): ")
verificacionEntero(duracion)
y=int(duracion)+1
muestreo=input("Ingrese el periodo de muestreo de la funci√≥n escal√≥n: ")
verificacionEntero(duracion)
escalon=vectorEscalon(x,y,muestreo)
tiempo = np.arange(0,y,muestreo)

#Graficaci√≥n
print("\nEl vector escal√≥n unitario tiene una duraci√≥n de:",duracion,"segundos e inicia en t=",n0,"segundos")
print("\nVector de tiempo= ",tiempo)
print("Vector escal√≥n= ",escalon)
plt.stem(tiempo, escalon, linefmt='deeppink', markerfmt='deeppink')
plt.title('ESCAL√ìN UNITARIO')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()

print("""\n
      Podemos evidenciar una entrada tipo bolus, ya que el escal√≥n en la vida real no es tan perfecto como se plantea.
      En un proceso fisiol√≥gico, aplicar una entrada de tipo escal√≥n perfecto es imposible,
      el cambio de '0' a '1' en el escal√≥n es poco realista.
      En la realidad, se tiene una se√±al con unos peque√±os cambios y pendientes, conocida como bolus.""")
```

Implemente una funci√≥n en Python que permita generar una funci√≥n rampa, definida en un
intervalo ùëõ1‚â§ùëõ0‚â§ùëõ2 con una pendiente m.

```python
#Hacemos la funci√≥n para el vector de rampa
def vectorRampa(t,duracion):
  rampa=[]
  i = 0  #Dado que en la rampa es 0 si n<n0, por eso iniciamos 'i' en cero.
  while i<duracion:
    if i<t:
      rampa.append(0)
      i += 1
    if i>=t:
      x=i-t
      rampa.append(x)
      i += 1
  return rampa

#Solicitamos el intervalo en el que desea la se√±al de rampa
n0=input("Ingrese el tiempo (en segundos) en el que desea que inicie la rampa: ")
verificacionEntero(n0)
x=int(n0)
duracion=input("Ingrese la duraci√≥n total de la rampa (en segundos): ")
verificacionEntero(duracion)
y=int(duracion)+1
rampa=vectorRampa(x,y)
tiempo = np.arange(0, y)

#Graficaci√≥n
print("\nEl vector rampa tiene una duraci√≥n de:",duracion,"segundos e inicia en t=",n0,"segundos")
print("\nVector de tiempo= ",tiempo)
print("Vector rampa= ",rampa)
plt.plot(tiempo, rampa, color='deeppink', linewidth=3)
plt.title('RAMPA')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```

```python
print(""" Por teor√≠a, se sabe que la funci√≥n delta de Dirac, conocida como impulso o Œ¥(n)
      es igual a 1 cuando ùëõ=0 y 0 en cualquier otro lugar.
      La expresi√≥n puede ser interpretada como una suma de impulsos en diferentes posiciones de ùëõ,
      con diferentes amplitudes.""")

print("""¬øC√ìMO SE DEBER√çA VER LA SE√ëAL?
      5ùõø(ùëõ+3)‚Üí Impulso de amplitud 5 en ùëõ = - 3
      3ùõø(ùëõ+2)‚Üí Impulso de amplitud 3 en ùëõ = - 2
      3Œ¥(ùëõ+1) ‚Üí Impulso de amplitud 3 en ùëõ = ‚àí 1
      3ùõø(ùëõ+0)‚Üí Impulso de amplitud 3 en ùëõ = 0
      1ùõø(ùëõ+0)‚Üí Impulso de amplitud 1 en ùëõ = 0

      Pero ojo, dado que tenemos dos t√©rminos en ùëõ=0, la amplitud total es 3+1=4.\n\n""")

#Hacemos una funci√≥n para crear el tiempo en donde vamos a graficar los impulsos. Como aqu√≠ ya sabemos que valores son, para
#reducir el c√≥digo yo voy a asumir que el valor de la izquierda siempre ser√° negativo y el de la derecha positivo
def vectorImpulso(a,b):
  impulso=[]
  i = 0  #Dado que en la rampa es 0 si n<n0, por eso iniciamos 'i' en cero.
  #Parte negativa del gr√°fico
  while a >= 0:
      impulso.append((a*-1))
      a -= 1
  #Parte positiva del gr√°fico
  i=1
  while i <= b:
    impulso.append(i)
    i += 1
  ceros=0
  while ceros <= 1:
    impulso += [0]  # Agregamos los '0' al final de la lista porque en ese instante el impulso no existe
    impulso.insert(0, 0)  # Agrega 0 al inicio de la lista
    ceros += 1
  return impulso

#Amplitud de los vectores impulso
def amplitudesImpulso(x):
  amplitudes=[]
  i=1
  y=x+1
  while i <= y:
    print("Ingrese la amplitud del vector",x-i)
    amplitud=int(input(" "))
    amplitudes.append(amplitud)
    i += 1
  ceros=0
  while ceros <= 1:
    amplitudes += [0]  # Agregamos los '0' al final de la lista porque en ese instante el impulso no existe
    amplitudes.insert(0, 0)  # Agrega 0 al inicio de la lista
    ceros += 1
  return amplitudes

#Obtenci√≥n de datos
extremoIzquierdo=input("Ingrese el tiempo (en segundos) en el que inician los impulsos: ")
verificacionEntero(extremoIzquierdo)
x=int(extremoIzquierdo)
extremoDerecho=input("Ingrese el tiempo (en segundos) en el que finalizan los impulsos: ")
verificacionEntero(extremoDerecho)
y=int(extremoDerecho)
impulse=vectorImpulso(x,y)
cantidadVectores=x+y
amplitude=amplitudesImpulso(cantidadVectores)


#Graficaci√≥n
print("\nVector de tiempo= ",impulse)
print("Vector de impulso= ",amplitude)
plt.scatter(impulse, amplitude, color='deeppink', linewidth=3)
plt.stem(impulse, amplitude, linefmt='deeppink', markerfmt='pink')
plt.title('IMPULSOS')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```

```python
#Creamos la funci√≥n para generar la secuencia
def secuenciaR(a,b,c):
  r=[]
  i=0
  while i<=a:
    r.append(i)
    i += 1
  while i > a and i<=b:
    r.append(i-5)
    i += 1
  while i > b and i<=c:
    r.append(i-10)
    i += 1
  return (r)

print("""\nSeg√∫n la literatura, se tiene que: La funci√≥n rampa tiene como primera derivada a la funci√≥n escal√≥n de Heaviside,
tambi√©n conocida como funci√≥n escal√≥n unitario, mientras que su segunda derivada
es la distribuci√≥n delta de Dirac (o distribuci√≥n Œ¥), tambi√©n conocida como impulso unitario
Link: https://doi.org/10.3390/appliedmath4020023""")

def derivadaRampa(a,b,c,x,y,z):
  escalon = []
  i = 0  #Dado que en el escal√≥n es 0 si n<n0, por eso iniciamos 'i' en cero.
  while i<=a:
    escalon.append(x)
    i += 1
  while i > a and i<=b:
    escalon.append(y)
    i += 1
  while i > b and i<=c:
    escalon.append(z)
    i += 1
  return (escalon)


vectores=secuenciaR(5,11,17)
tiempo = np.arange(0, 18)
derivada=derivadaRampa(5,11,17,5,6,7)


#Graficaci√≥n
print("\nVector de tiempo = ",tiempo)
print("Vector de rampa = ",vectores)
print("Vector derivada (escal√≥n unitario) = ",derivada,"\n")
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.stem(tiempo, vectores,linefmt='deeppink', markerfmt='deeppink')
plt.title('Rampas')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid(True)
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.plot(tiempo, derivada,color='deeppink')
plt.stem([0,17], [5,7],linefmt='deeppink', markerfmt='deeppink')
plt.title('Derivada de la rampa (escal√≥n unitario)')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()

print("""\nLa secuencia dada presenta una rampa que se reinicia cada 5 unidades de tiempo (es una funci√≥n A TRAMOS), incrementando su valor en 1 unidad en cada ciclo.
Esta secuencia, aunque presenta caracter√≠sticas similares a una funci√≥n rampa continua,
presenta discontinuidades en su primera derivada debido a los reinicios peri√≥dicos.
En la gr√°fica original podemos observar una serie de segmentos lineales con pendiente positiva,
cada uno iniciado en un valor de distinto cada 5 unidades de tiempo. Al final de cada segmento,
la gr√°fica vuelve a cero para iniciar un nuevo segmento con aumento de una unidad.
En una rampa continua, la derivada es una funci√≥n escal√≥n unitario constante.
En la secuencia dada, la derivada son 3 escalones unitarios con amplitud diferente en los puntos donde la rampa inicia un nuevo ciclo,
ya que hay un salto instant√°neo en el valor de la funci√≥n.
""")
```
