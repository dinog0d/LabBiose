Obtenemos en nuestro directorio actual, las carpetas de control y de parkinson con los datos a usar, vamos a obtener las listas con los nombres de los archivos para poder realizar el proceso solicitado m√°s adelante.

```python
import os
import scipy.io as sio
import numpy as np

directorio_actual = os.getcwd()

# Creamos una funci√≥n que va a recibir este directorio actual y la carpeta de la que deseemos extraer los datos

def cargar_senales(directorio, subdirectorio):
    # Listamos los archivos en el subdirectorio
    archivos = os.listdir(os.path.join(directorio, subdirectorio))
    senales = []

    # Cargamos y procesamos los archivos
    for archivo in archivos:
        ruta_archivo = os.path.join(directorio, subdirectorio, archivo)
        print(ruta_archivo) # Imprimimos cada ruta del archivo para visualizar que es√©n cargando todos los archivos .mat
        # Cargar archivo
        data = sio.loadmat(ruta_archivo)["data"]
        senales.append(data)

    return senales

# Cargar se√±ales de control
control = cargar_senales(directorio_actual, 'control')

# Cargar se√±ales de Parkinson
parkinson = cargar_senales(directorio_actual, 'parkinson')
```

**1. Implemente una funci√≥n que reciba una se√±al de m√∫ltiples canales y √©pocas y calcule la Energ√≠a de promedio de cada canal.**

```python
def calcular_energia_promedio(senal):
  energia_promedio = []
  for canal in range(senal.shape[0]):
    energia_canal = np.mean(np.sum(senal[canal, :, :] ** 2, axis=0))
    energia_promedio.append(energia_canal)
  return energia_promedio

calcular_energia_promedio(control[0]) # Probamos la funci√≥n el control 18
```

**2. Calcule la energ√≠a de cada canal promediada por √©pocas para cada sujeto, esto para ambos grupos poblacionales. Guarde esta informaci√≥n en un DataFrame de columnas ‚Äòcanal‚Äô y filas ‚Äò#sujeto‚Äô con los valores de energ√≠a calculados, un DataFrame para cada grupo poblacional.**  

```python
import pandas as pd

def calcular_energia_promedio_por_sujeto(senales):
  energia_promedio_sujetos = []
  for i, senal in enumerate(senales):
    energia_promedio_canal = calcular_energia_promedio(senal)
    energia_promedio_sujetos.append(energia_promedio_canal)
  df = pd.DataFrame(energia_promedio_sujetos, columns=[f"Canal {i+1}" for i in range(len(energia_promedio_sujetos[0]))])
  df.index.name = 'Sujeto'
  return df

# Calcular la energ√≠a promedio por canal para cada sujeto en el grupo control
df_control = calcular_energia_promedio_por_sujeto(control)

# Calcular la energ√≠a promedio por canal para cada sujeto en el grupo parkinson
df_parkinson = calcular_energia_promedio_por_sujeto(parkinson)

# Reindexar el DataFrame de control
df_control.index = df_control.index + 1

# Reindexar el DataFrame de parkinson
df_parkinson.index = df_parkinson.index + 1
```

**3. Determine si existe diferencia estad√≠stica entre canales de cada grupo de sujetos a trav√©s de una pruebat.Compruebe los supuestos necesarios para realizar una prueba t, esto es: Normalidad de la variable, independencia (se asume que los grupos son independientes), y homocedasticidad (use una prueba de Levene), finalmente realice la prueba t para determinar si existen diferencias entre los canales entre grupos de sujetos. De no cumplirse los requisitos, realice entonces un an√°lisis no param√©trico (prueba U de Mann-Whitney). Este numeral tiene como objetivo identificar los canales que entregan informaci√≥n diferencial entre pacientes Sanos y con enfermedad de Parkinson**

```python
from scipy.stats import shapiro, levene, ttest_ind, mannwhitneyu
import pandas as pd

# Funci√≥n para comprobar normalidad con 0.05
def prueba_normalidad(df):
    _, p_shapiro = shapiro(df)
    return p_shapiro, p_shapiro > 0.05

# Funci√≥n para comprobar homocedasticidad con 0.05
def prueba_homocedasticidad(df_1,df_2):
    _, p_levene = levene(df_1,df_2)
    return p_levene, p_levene > 0.05

# Funci√≥n para realizar la prueba t para prueba parametrica
def prueba_t(df_1, df_2):
    _, p_ttest = ttest_ind(df_1, df_2)
    return p_ttest, p_ttest > 0.05

# Funci√≥n para realizar la prueba U de Mann-Whitney para prueba no-parametrica
def prueba_mann_whitney(df_1, df_2):
    _, p_mannwhitney = mannwhitneyu(df_1, df_2)
    return p_mannwhitney, p_mannwhitney > 0.05
```

### *AN√ÅLISIS DE PRUEBA PARAM√âTRICA* ###


La prueba t-student es una t√©cnica empleada para identificar si las medias entre dos grupos de estudio son similares o diferentes [1], en nuestro caso se desea determinar si las medias entre los pacientes de Parkinson y Pacientes de control sanos son iguales o distintas. Para ejecutar esta prueba es necesario rectificar los siguientes supuestos:

‚Ä¢	Independencia de los datos implementados, donde las medidas de observaci√≥n no se afectan entre s√≠, para los datos adquiridos en esta pr√°ctica se asume su independencia.

‚Ä¢	Los datos siguen una distribuci√≥n normal, conocida.

‚Ä¢	La varianza de las dos muestras analizadas es igual.




### **Prueba** **de normalidad**
Para comprobar la distribuci√≥n de los datos se ejecut√≥ una prueba Shapiro la cual puede ser realizada en grupos con un tama√±o m√°ximo de 50 muestras [2]. Debido a que en esta pr√°ctica se realiza un an√°lisis en 16 canales (entre pacientes sanos y diagnosticados con Parkinson), este tipo de t√©cnica puede ser aplicada sobre los datos, para ello se plantea a continuaci√≥n las siguientes hip√≥tesis de normalidad estableciendo como criterio un Œ±=0.05:

- **Ho:**  Los datos tienen una distribuci√≥n de forma normal, conocida.

- **Ha:** Los datos no cuentan una distribuci√≥n normal ni conocida.



```python
for canal in df_control.columns[:]: #Se recorren todas las columnas de los canales
        normalidad = prueba_normalidad(df_control[canal]) #Se llama la prueba de normalidad
        if normalidad[1]:
          print(f'Para la prueba de normalidad del {canal} de Control tenemos un p:{normalidad[0]:.3f} por consecuente se acepta la hip√≥tesis nula indicando que los grupos cuentan con una distribuci√≥n normal conocida.\n')
        else:
          print(f'Para la prueba de normalidad del {canal} de Control tenemos un p:{normalidad[0]:.3f} que es menor a 0.05 por consecuente se rechaza la hip√≥tesis nula y respalda la hip√≥tesis alternativa indicando que los grupos no cuentan con una distribuci√≥n normal conocida.\n')
```

```python
for canal in df_parkinson.columns[:]: #Se recorren todas las columnas de los canales
        normalidad = prueba_normalidad(df_parkinson[canal]) #Se llama la prueba de normalidad
        if normalidad[1]:
          print(f'Para la prueba de normalidad del {canal} de Parkinson tenemos un p:{normalidad[0]:.3f} por consecuente se acepta la hip√≥tesis nula indicando que los grupos cuentan con una distribuci√≥n normal conocida.\n')
        else:
          print(f'Para la prueba de normalidad del {canal} de Parkinson tenemos un p:{normalidad[0]:.3f} que es menor a 0.05 por consecuente se rechaza la hip√≥tesis nula y respalda la hip√≥tesis alternativa indicando que los grupos no cuentan con una distribuci√≥n normal conocida.\n')
```

Ahora bien, al aplicar la prueba Shapiro en cada canal obtenido de los pacientes de control sanos versus los canales obtenidos en los pacientes con Parkinson se observ√≥ una p < Œ± por consecuente se rechaza la hip√≥tesis nula y respalda la hip√≥tesis alternativa indicando que los grupos no cuentan con una distribuci√≥n normal conocida.

### **Prueba de homocedasticidad** ###

Con el fin de analizar la similitud de las varianzas entre cada grupo de datos se ejecut√≥ una prueba de Levene, la cual permite corroborar o descartar el hecho de que las muestras proceden de una poblaci√≥n con igual varianza y que sus diferencias pueden o no provenir de elementos aleatorios. Para desarrollar este m√©todo se estipularon las siguientes hip√≥tesis en cada uno de los canales de datos:
- **Ho:** Los grupos cuentan con igualdad de varianzas œÉ1= œÉ2 =‚Ä¶= œÉn
- **Ha:** Al menos una de las varianzas de los grupos es diferente.


```python
for canal in df_control.columns:
        # Extraer los datos para el canal actual
        datos_control = df_control[canal]
        datos_parkinson = df_parkinson[canal]

        # Realizar la prueba de Levene
        homocedasticidad = prueba_homocedasticidad(datos_control, datos_parkinson)
        if homocedasticidad[1]:
            print(f"Para la prueba de homocedasticidad del {canal} de Control y Parkinson tenemos un p:{homocedasticidad[0]:.3f} por consecuente se acepta la hip√≥tesis nula un lo cual conlleva a descartar la hip√≥tesis alternativa.\n")
        else:
            print(f"Para la prueba de homocedasticidad del {canal} de Control y Parkinson tenemos un p:{homocedasticidad[0]:.3f} por consecuente se descarta la hip√≥tesis nula un lo cual conlleva a soportar la hip√≥tesis alternativa.\n")
```

Al trabajar esta prueba con un Œ±=0.05 se obtuvo un p > Œ± lo cual conlleva a descartar la hip√≥tesis alternativa ya que se cuenta con suficiente informaci√≥n para respaldar la hip√≥tesis nula, es as√≠ como se se√±ala que los datos tienen homocedasticidad, es decir igualdad de varianza.
No obstante, al no cumplirse el criterio de normalidad el cual se establece como el criterio principal para ejecutar la prueba T-student, se descarta el desarrollo de un an√°lisis param√©trico de los datos y se conlleva a realizar la prueba No param√©trica U de Mann Whitney.


### **AN√ÅLISIS DE PRUEBA NO PARAM√âTRICA** ###
Como consecuencia de la no normalidad de las muestras, se aplica una prueba U de Mann Whitney, la cual es un m√©todo alternativo de la prueba t de muestras independientes. El objetivo de esta t√©cnica es determinar la existencia de disimilitudes entre las variables dependientes de 2 grupos [https://datatab.es/tutorial/mann-whitney-u-test] (distinciones en la tendencia central), siendo para esta pr√°ctica utilizada para determinar diferencias en la energ√≠a de la se√±al EEG entre los grupos de control de pacientes sanos y con Parkinson.  Por medio de lo anterior, se tomaron en cuanta las siguientes hip√≥tesis:
- **Ho:** No existe diferencia en la tendencia central (mediana) entre los canales de energ√≠a de los pacientes de control sanos y diagnosticados con Parkinson.
- **Ha:** Existe diferencia en la tendencia central de los datos entre los grupos de control de pacientes sanos y diagnosticados con Parkinson.


```python
for canal in df_control.columns:
        # Extraer los datos para el canal actual
        datos_control = df_control[canal]
        datos_parkinson = df_parkinson[canal]

        # Realizar la prueba de Levene
        mann_whitney = prueba_mann_whitney(datos_control, datos_parkinson)
        if mann_whitney[1]:
            print(f"Para la prueba de U-mann Whitney del {canal} de Control y Parkinson tenemos un p:{mann_whitney[0]:.3f} por consecuente se acepta la hip√≥tesis nula un lo cual conlleva a descartar la hip√≥tesis alternativa.\n")
        else:
            print(f"Para la prueba de U-mann Whitney del {canal} de Control y Parkinson tenemos un p:{mann_whitney[0]:.3f} por consecuente se descarta la hip√≥tesis nula un lo cual conlleva a soportar la hip√≥tesis alternativa.\n")
```

Para el an√°lisis anterior se implement√≥ un Œ±=0.05 y se obtuvo un valor p > Œ±, por consecuente se respalda la hip√≥tesis nula y se rechaza Ha, es as√≠ como los datos NO cuentan con diferencias en su tendencia central, por lo tanto, no es posible identificar diferencias en los canales de energ√≠a entre los grupos de pacientes de control sanos y diagnosticados con Alzheimer.
